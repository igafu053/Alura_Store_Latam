### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.


- Calculando ingresos para Tienda 1:

```python
# Paso 1: Calcular ingresos para tienda_1.csv
print("Calculando ingresos para Tienda 1: ")
tienda = pd.read_csv(url)
total_tienda = tienda['Precio'].sum().round(2)  # Total vendido en Tienda 1
print(f"\nTotal vendido en Tienda 1: $ {total_tienda:,.0f}")
```

- Calculando ingresos para Tienda 2:

```python
# Paso 2: Calcular ingresos para tienda_2.csv
print("Calculando ingresos para Tienda 2: ")
tienda2 = pd.read_csv(url2)
total_tienda2 = tienda2['Precio'].sum().round(2)  # Total vendido en Tienda 2
print(f"\nTotal vendido en Tienda 2: $ {total_tienda2:,.0f}")
```

- Calculando ingresos para Tienda 3:

```python
# Paso 3: Calcular ingresos para tienda_3.csv
print("Calculando ingresos para Tienda 3: ")
tienda3 = pd.read_csv(url3)
total_tienda3 = tienda3['Precio'].sum().round(2)  # Total vendido en Tienda 3
print(f"\nTotal vendido en Tienda 3: $ {total_tienda3:,.0f}")
```

- Calculando ingresos para Tienda 4:

```python
# Paso 4: Calcular ingresos para tienda_4.csv
print("Calculando ingresos para Tienda 4: ")
tienda4 = pd.read_csv(url4)
total_tienda4 = tienda4['Precio'].sum().round(2)  # Total vendido en Tienda 4
print(f"\nTotal vendido en Tienda 4: $ {total_tienda4:,.0f}")
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

- Calculando productos por categoría para Tienda 1:

```python
# Diccionario para guardar las categorías más populares de cada tienda
categorias_mas_populares = {}

# Paso 1: Calcular productos vendidos por categoría para tienda_1.csv
print("Calculando productos por categoría para Tienda 1: ")
tienda = pd.read_csv(url)
productos_tienda = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
categorias_mas_populares['Tienda 1'] = productos_tienda.index[0]  # Guardar la categoría más popular
print("\nProductos vendidos por categoría en Tienda 1:\n")
print(productos_tienda)
```

- Calculando productos por categoría para Tienda 2:

```python
# Paso 2: Calcular productos vendidos por categoría para tienda_2.csv
print("Calculando productos por categoría para Tienda 2: ")
tienda2 = pd.read_csv(url2)
productos_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
categorias_mas_populares['Tienda 2'] = productos_tienda2.index[0]  # Guardar la categoría más popular
print("\nProductos vendidos por categoría en Tienda 2:\n")
print(productos_tienda2)
```

- Calculando productos por categoría para Tienda 3:

```python
# Paso 3: Calcular productos vendidos por categoría para tienda_3.csv
print("Calculando productos por categoría para Tienda 3: ")
tienda3 = pd.read_csv(url3)
productos_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
categorias_mas_populares['Tienda 3'] = productos_tienda3.index[0]  # Guardar la categoría más popular
print("\nProductos vendidos por categoría en Tienda 3:\n")
print(productos_tienda3)
```

- Calculando productos por categoría para Tienda 4:

```python
# Paso 4: Calcular productos vendidos por categoría para tienda_4.csv
print("Calculando productos por categoría para Tienda 4: ")
tienda4 = pd.read_csv(url4)
productos_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)
categorias_mas_populares['Tienda 4'] = productos_tienda4.index[0]  # Guardar la categoría más popular
print("\nProductos vendidos por categoría en Tienda 4:\n")
print(productos_tienda4)
```

- Resumen de Categorías Más Populares

```python
# Paso 5: Resumen de las categorías más populares
print("=== Resumen de Categorías Más Populares ===\n")
for tienda, categoria in categorias_mas_populares.items():
    print(f"* Categoría más popular en {tienda}: {categoria}")
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

- Analizando productos vendidos en Tienda 1:

```python
# Diccionarios para guardar los productos más y menos vendidos
mas_vendidos = {}
menos_vendidos = {}

# Paso 1: Calcular productos vendidos para tienda_1.csv
print("Analizando productos vendidos en Tienda 1: \n")
tienda = pd.read_csv(url)
ventas_tienda1 = tienda.groupby('Producto').size().sort_values(ascending=False)
producto_mas_vendido_t1 = ventas_tienda1.index[0]  # Producto con más ventas
cantidad_mas_vendido_t1 = ventas_tienda1.iloc[0]   # Cantidad de ventas del más vendido
producto_menos_vendido_t1 = ventas_tienda1.index[-1]  # Producto con menos ventas
cantidad_menos_vendido_t1 = ventas_tienda1.iloc[-1]   # Cantidad de ventas del menos vendido
mas_vendidos['Tienda 1'] = (producto_mas_vendido_t1, cantidad_mas_vendido_t1)
menos_vendidos['Tienda 1'] = (producto_menos_vendido_t1, cantidad_menos_vendido_t1)
print(f"* Producto más vendido en Tienda 1: {producto_mas_vendido_t1} -> ({cantidad_mas_vendido_t1} ventas)")
print(f"* Producto menos vendido en Tienda 1: {producto_menos_vendido_t1} -> ({cantidad_menos_vendido_t1} ventas)")
```

- Analizando productos vendidos en Tienda 2:

```python
# Paso 2: Calcular productos vendidos para tienda_2.csv
print("Analizando productos vendidos en Tienda 2: \n")
tienda2 = pd.read_csv(url2)
ventas_tienda2 = tienda2.groupby('Producto').size().sort_values(ascending=False)
producto_mas_vendido_t2 = ventas_tienda2.index[0]
cantidad_mas_vendido_t2 = ventas_tienda2.iloc[0]
producto_menos_vendido_t2 = ventas_tienda2.index[-1]
cantidad_menos_vendido_t2 = ventas_tienda2.iloc[-1]
mas_vendidos['Tienda 2'] = (producto_mas_vendido_t2, cantidad_mas_vendido_t2)
menos_vendidos['Tienda 2'] = (producto_menos_vendido_t2, cantidad_menos_vendido_t2)
print(f"* Producto más vendido en Tienda 2: {producto_mas_vendido_t2} -> ({cantidad_mas_vendido_t2} ventas)")
print(f"* Producto menos vendido en Tienda 2: {producto_menos_vendido_t2} -> ({cantidad_menos_vendido_t2} ventas)")
```

- Analizando productos vendidos en Tienda 3:

```python
# Paso 3: Calcular productos vendidos para tienda_3.csv
print("Analizando productos vendidos en Tienda 3: \n")
tienda3 = pd.read_csv(url3)
ventas_tienda3 = tienda3.groupby('Producto').size().sort_values(ascending=False)
producto_mas_vendido_t3 = ventas_tienda3.index[0]
cantidad_mas_vendido_t3 = ventas_tienda3.iloc[0]
producto_menos_vendido_t3 = ventas_tienda3.index[-1]
cantidad_menos_vendido_t3 = ventas_tienda3.iloc[-1]
mas_vendidos['Tienda 3'] = (producto_mas_vendido_t3, cantidad_mas_vendido_t3)
menos_vendidos['Tienda 3'] = (producto_menos_vendido_t3, cantidad_menos_vendido_t3)
print(f"* Producto más vendido en Tienda 3: {producto_mas_vendido_t3} -> ({cantidad_mas_vendido_t3} ventas)")
print(f"* Producto menos vendido en Tienda 3: {producto_menos_vendido_t3} -> ({cantidad_menos_vendido_t3} ventas)")
```

- Analizando productos vendidos en Tienda 4:

```python
# Paso 4: Calcular productos vendidos para tienda_4.csv
print("Analizando productos vendidos en Tienda 4: \n")
tienda4 = pd.read_csv(url4)
ventas_tienda4 = tienda4.groupby('Producto').size().sort_values(ascending=False)
producto_mas_vendido_t4 = ventas_tienda4.index[0]
cantidad_mas_vendido_t4 = ventas_tienda4.iloc[0]
producto_menos_vendido_t4 = ventas_tienda4.index[-1]
cantidad_menos_vendido_t4 = ventas_tienda4.iloc[-1]
mas_vendidos['Tienda 4'] = (producto_mas_vendido_t4, cantidad_mas_vendido_t4)
menos_vendidos['Tienda 4'] = (producto_menos_vendido_t4, cantidad_menos_vendido_t4)
print(f"* Producto más vendido en Tienda 4: {producto_mas_vendido_t4} -> ({cantidad_mas_vendido_t4} ventas)")
print(f"* Producto menos vendido en Tienda 4: {producto_menos_vendido_t4} -> ({cantidad_menos_vendido_t4} ventas)")
```

- Resumen de Productos Más y Menos Vendidos

```python
# Paso 5: Resumen de productos más y menos vendidos
print("=== Resumen de Productos Más y Menos Vendidos ===\n")
print("Productos más vendidos por tienda: \n")
for tienda, (producto, cantidad) in mas_vendidos.items():
    print(f"* {tienda}: {producto} -> ({cantidad} ventas)")
print()
print("Productos menos vendidos por tienda: \n")
for tienda, (producto, cantidad) in menos_vendidos.items():
    print(f"* {tienda}: {producto} -> ({cantidad} ventas)")
```
