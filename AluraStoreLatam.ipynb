#1. Análisis de facturación

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# Paso 1: Calcular ingresos para tienda_1.csv
print("Calculando ingresos para Tienda 1: ")
tienda = pd.read_csv(url)
total_tienda = tienda['Precio'].sum().round(2)  # Total vendido en Tienda 1
print(f"\nTotal vendido en Tienda 1: $ {total_tienda:,.0f}")
```

# 2. Ventas por categoría

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URLs para las bases de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar bases de datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular el total de ventas para cada tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [
    tienda['Precio'].sum().round(2),
    tienda2['Precio'].sum().round(2),
    tienda3['Precio'].sum().round(2),
    tienda4['Precio'].sum().round(2)
]

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Total de ventas por tienda', fontsize=14, pad=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Ingresos por tiendas ($)', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Agregar nivel del total de las ventas sobre las barras del gr+afico
for i, v in enumerate(ingresos):
    plt.text(i, v + 1000000, f'${v:,.0f}', ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('revenue_by_store.png')
plt.show()
plt.close()
```

- Gráfico ventas por categoría en cada tienda

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URLs para las bases de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar bases de datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Definir categorías (para asegurarse de que todas están incluidas, incluso si hay cero en algunas tiendas)
categories = ['Muebles', 'Electrónicos', 'Juguetes', 'Electrodomésticos',
              'Deportes y diversión', 'Instrumentos musicales', 'Libros',
              'Artículos para el hogar']

# Calcular las ventas por categoria para cada tienda
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
category_counts = {
    'Tienda 1': tienda.groupby('Categoría del Producto').size().reindex(categories, fill_value=0),
    'Tienda 2': tienda2.groupby('Categoría del Producto').size().reindex(categories, fill_value=0),
    'Tienda 3': tienda3.groupby('Categoría del Producto').size().reindex(categories, fill_value=0),
    'Tienda 4': tienda4.groupby('Categoría del Producto').size().reindex(categories, fill_value=0)
}

# Crear gráfico de barras apiladas
plt.figure(figsize=(12, 8))
bottom = [0] * len(stores)  # Inicializar la parte inferior para apilar
for category in categories:
    counts = [category_counts[store][category] for store in stores]
    plt.bar(stores, counts, bottom=bottom, label=category)
    bottom = [bottom[i] + counts[i] for i in range(len(bottom))]

plt.title('Ventas por categoría para cada tienda', fontsize=14, pad=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Cantidad de Ventas', fontsize=12)
plt.legend(title='Categoría del Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('category_sales_by_store.png')
plt.show()
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# URLs para las bases de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar las bases de datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular los productos más y menos vendidos para cada tienda
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
most_sold = []
least_sold = []
most_sold_products = []
least_sold_products = []

# Tienda 1
ventas_tienda1 = tienda.groupby('Producto').size().sort_values(ascending=False)
most_sold.append(ventas_tienda1.iloc[0])
least_sold.append(ventas_tienda1.iloc[-1])
most_sold_products.append(ventas_tienda1.index[0])
least_sold_products.append(ventas_tienda1.index[-1])

# Tienda 2
ventas_tienda2 = tienda2.groupby('Producto').size().sort_values(ascending=False)
most_sold.append(ventas_tienda2.iloc[0])
least_sold.append(ventas_tienda2.iloc[-1])
most_sold_products.append(ventas_tienda2.index[0])
least_sold_products.append(ventas_tienda2.index[-1])

# Tienda 3
ventas_tienda3 = tienda3.groupby('Producto').size().sort_values(ascending=False)
most_sold.append(ventas_tienda3.iloc[0])
least_sold.append(ventas_tienda3.iloc[-1])
most_sold_products.append(ventas_tienda3.index[0])
least_sold_products.append(ventas_tienda3.index[-1])

# Tienda 4
ventas_tienda4 = tienda4.groupby('Producto').size().sort_values(ascending=False)
most_sold.append(ventas_tienda4.iloc[0])
least_sold.append(ventas_tienda4.iloc[-1])
most_sold_products.append(ventas_tienda4.index[0])
least_sold_products.append(ventas_tienda4.index[-1])

# Crear diagrama de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(stores, most_sold, color='green', marker='^', s=100, label='Más vendidos')
plt.scatter(stores, least_sold, color='red', marker='v', s=100, label='Menos vendidos')

# Agregar nombres de productos como anotaciones
for i, (most_prod, least_prod, most_count, least_count) in enumerate(zip(most_sold_products, least_sold_products, most_sold, least_sold)):
    plt.text(i, most_count + 1, most_prod, ha='center', fontsize=8, color='green')
    plt.text(i, least_count - 2, least_prod, ha='center', fontsize=8, color='red')

plt.title('Productos más y menos vendidos por tienda', fontsize=14, pad=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Cantidad de ventas', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.ylim(25, 70)  # Establecer los límites del eje y para centrarse en el rango de ventas
plt.tight_layout()
plt.savefig('product_sales_by_store.png')
plt.show()
plt.close()
```

```python
import pandas as pd

# Cargar las bases de datos desde las URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Combinar todas las bases de datos en una sola
df = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

# Limpiar datos: convertir 'lat' y 'lon' a numéricos y eliminar filas con datos faltantes
df['lat'] = pd.to_numeric(df['lat'], errors='coerce')
df['lon'] = pd.to_numeric(df['lon'], errors='coerce')
df = df.dropna(subset=['lat', 'lon', 'Precio', 'Lugar de Compra', 'Calificación'])
```

**1.- Mapear las ventas de cada tienda y analizar la distribución geográfica**

```python
# Paso 1: Mapear ventas por ubicación
# Contar cuántas ventas hay por ciudad
ventas_por_ciudad = df['Lugar de Compra'].value_counts()

# Resumen de ventas totales (suma de Precio) por ciudad
ventas_totales_por_ciudad = df.groupby('Lugar de Compra')['Precio'].sum().sort_values(ascending=False)
# Formato 1: Separadores de miles
ventas_formateadas = ventas_totales_por_ciudad.map('{:,.0f}'.format)
# Formato 2: En millones
ventas_millones = ventas_totales_por_ciudad / 1_000_000
```

**2.- Generar gráficos de dispersión para visualizar datos**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 2: Gráfico de dispersión
plt.figure(figsize=(10, 6))  # Tamaño del gráfico
sns.scatterplot(x='lon', y='lat', size='Precio', sizes=(20, 200), hue='Lugar de Compra', data=df)
plt.title('Distribución geográfica de ventas')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Leyenda fuera del gráfico
plt.tight_layout()
plt.show()
```

**3.- Explorar el rendimiento de las tiendas por región**

```python
# Paso 3: Resumen de rendimiento por ciudad
city_stats = df.groupby('Lugar de Compra').agg({
    'Precio': 'sum',           # Total de ventas
    'Calificación': 'mean',    # Promedio de calificaciones
    'Producto': 'count'        # Número de ventas
}).reset_index()

city_stats.columns = ['Ciudad', 'Ventas Totales', 'Calificación Promedio', 'Cantidad de Ventas']
# Convertir Ventas Totales a millones y formatear Calificación Promedio
city_stats['Ventas Totales (Millones)'] = city_stats['Ventas Totales']/ 1_000_000
city_stats['Calificación Promedio'] = city_stats['Calificación Promedio'].round(2)
# Ordenar las ciudades por Ventas Totales (Millones) de mayor a menor
city_stats = city_stats.sort_values(by='Ventas Totales (Millones)', ascending=False)
# Ordenar las ciudades por Calificación Promedio de mayor a menor
city_stats = city_stats.sort_values(by='Calificación Promedio', ascending=False)
```

```python
# Gráfico de barras horizontales: Ventas totales (en millones), de mayor a menor
plt.figure(figsize=(10, 8))  # Ajustar tamaño para tantas ciudades
palette = sns.color_palette("tab20", n_colors=len(city_stats))
sns.barplot(x='Ciudad', y='Ventas Totales (Millones)', hue='Ciudad', palette=palette, data=city_stats, legend=False )
plt.title('Ventas Totales por Ciudad (en Millones)')
plt.xticks(rotation=45)
plt.xlabel('Ciudades')
plt.ylabel('Ventas Totales (Millones)')
plt.tight_layout()
plt.show()
```
