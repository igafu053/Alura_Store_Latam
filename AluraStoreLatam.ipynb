#1. Análisis de facturación

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.


- Calculando ingresos para Tienda 1:

```python
# Paso 1: Calcular ingresos para tienda_1.csv
print("Calculando ingresos para Tienda 1: ")
tienda = pd.read_csv(url)
total_tienda = tienda['Precio'].sum().round(2)  # Total vendido en Tienda 1
print(f"\nTotal vendido en Tienda 1: $ {total_tienda:,.0f}")
```

```python
# Paso 3: Calcular ingresos para tienda_3.csv
print("Calculando ingresos para Tienda 3: ")
tienda3 = pd.read_csv(url3)
total_tienda3 = tienda3['Precio'].sum().round(2)  # Total vendido en Tienda 3
print(f"\nTotal vendido en Tienda 3: $ {total_tienda3:,.0f}")
```

- Calculando ingresos para Tienda 4:

```python
# Paso 4: Calcular ingresos para tienda_4.csv
print("Calculando ingresos para Tienda 4: ")
tienda4 = pd.read_csv(url4)
total_tienda4 = tienda4['Precio'].sum().round(2)  # Total vendido en Tienda 4
print(f"\nTotal vendido en Tienda 4: $ {total_tienda4:,.0f}")
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

- Calculando productos por categoría para Tienda 1:

```python
# Diccionario para guardar las categorías más populares de cada tienda
categorias_mas_populares = {}

# Paso 1: Calcular productos vendidos por categoría para tienda_1.csv
print("Calculando productos por categoría para Tienda 1: ")
tienda = pd.read_csv(url)
productos_tienda = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
categorias_mas_populares['Tienda 1'] = productos_tienda.index[0]  # Guardar la categoría más popular
print("\nProductos vendidos por categoría en Tienda 1:\n")
print(productos_tienda)
```

- Calculando productos por categoría para Tienda 3:

```python
# Paso 3: Calcular productos vendidos por categoría para tienda_3.csv
print("Calculando productos por categoría para Tienda 3: ")
tienda3 = pd.read_csv(url3)
productos_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
categorias_mas_populares['Tienda 3'] = productos_tienda3.index[0]  # Guardar la categoría más popular
print("\nProductos vendidos por categoría en Tienda 3:\n")
print(productos_tienda3)
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
# Diccionarios para guardar los productos más y menos vendidos
mas_vendidos = {}
menos_vendidos = {}

# Paso 1: Calcular productos vendidos para tienda_1.csv
print("Analizando productos vendidos en Tienda 1: \n")
tienda = pd.read_csv(url)
ventas_tienda1 = tienda.groupby('Producto').size().sort_values(ascending=False)
producto_mas_vendido_t1 = ventas_tienda1.index[0]  # Producto con más ventas
cantidad_mas_vendido_t1 = ventas_tienda1.iloc[0]   # Cantidad de ventas del más vendido
producto_menos_vendido_t1 = ventas_tienda1.index[-1]  # Producto con menos ventas
cantidad_menos_vendido_t1 = ventas_tienda1.iloc[-1]   # Cantidad de ventas del menos vendido
mas_vendidos['Tienda 1'] = (producto_mas_vendido_t1, cantidad_mas_vendido_t1)
menos_vendidos['Tienda 1'] = (producto_menos_vendido_t1, cantidad_menos_vendido_t1)
print(f"* Producto más vendido en Tienda 1: {producto_mas_vendido_t1} -> ({cantidad_mas_vendido_t1} ventas)")
print(f"* Producto menos vendido en Tienda 1: {producto_menos_vendido_t1} -> ({cantidad_menos_vendido_t1} ventas)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs para las bases de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar bases de datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular el total de ventas para cada tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [
    tienda['Precio'].sum().round(2),
    tienda2['Precio'].sum().round(2),
    tienda3['Precio'].sum().round(2),
    tienda4['Precio'].sum().round(2)
]

# Crear gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Total de ventas por tienda', fontsize=14, pad=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Ingresos por tiendas ($)', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Agregar nivel del total de las ventas sobre las barras del gr+afico
for i, v in enumerate(ingresos):
    plt.text(i, v + 1000000, f'${v:,.0f}', ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('revenue_by_store.png')
```

- Gráfico ventas por categoría en cada tienda

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs para las bases de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar bases de datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Definir categorías (para asegurarse de que todas están incluidas, incluso si hay cero en algunas tiendas)
categories = ['Muebles', 'Electrónicos', 'Juguetes', 'Electrodomésticos',
              'Deportes y diversión', 'Instrumentos musicales', 'Libros',
              'Artículos para el hogar']

# Calcular las ventas por categoria para cada tienda
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
category_counts = {
    'Tienda 1': tienda.groupby('Categoría del Producto').size().reindex(categories, fill_value=0),
    'Tienda 2': tienda2.groupby('Categoría del Producto').size().reindex(categories, fill_value=0),
    'Tienda 3': tienda3.groupby('Categoría del Producto').size().reindex(categories, fill_value=0),
    'Tienda 4': tienda4.groupby('Categoría del Producto').size().reindex(categories, fill_value=0)
}

# Crear gráfico de barras apiladas
plt.figure(figsize=(12, 8))
bottom = [0] * len(stores)  # Inicializar la parte inferior para apilar
for category in categories:
    counts = [category_counts[store][category] for store in stores]
    plt.bar(stores, counts, bottom=bottom, label=category)
    bottom = [bottom[i] + counts[i] for i in range(len(bottom))]

plt.title('Ventas por categoría para cada tienda', fontsize=14, pad=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Cantidad de Ventas', fontsize=12)
plt.legend(title='Categoría del Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('category_sales_by_store.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs para las bases de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar las bases de datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular la valoración media de cada tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [
    tienda['Calificación'].mean().round(2),
    tienda2['Calificación'].mean().round(2),
    tienda3['Calificación'].mean().round(2),
    tienda4['Calificación'].mean().round(2)
]

# Crear gráfico de líneas
plt.figure(figsize=(10, 6))
plt.plot(tiendas, calificaciones, marker='o', color='purple', linewidth=2, markersize=8)
plt.title('Valoración media de los clientes por tienda', fontsize=14, pad=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Valoración media (1-5)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Añadir etiquetas de valor encima de cada punto
for i, v in enumerate(calificaciones):
    plt.text(i, v + 0.02, f'{v}', ha='center', fontsize=10)

plt.ylim(3.8, 4.2)  # Establecer los limites del eje y para centrarse en el rango de ratings
plt.tight_layout()
plt.savefig('average_rating_by_store.png')
```

- Graficos productos más y menos vendidos por tienda

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs para las bases de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar las bases de datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular los productos más y menos vendidos para cada tienda
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
most_sold = []
least_sold = []
most_sold_products = []
least_sold_products = []

# Tienda 1
ventas_tienda1 = tienda.groupby('Producto').size().sort_values(ascending=False)
most_sold.append(ventas_tienda1.iloc[0])
least_sold.append(ventas_tienda1.iloc[-1])
most_sold_products.append(ventas_tienda1.index[0])
least_sold_products.append(ventas_tienda1.index[-1])

# Tienda 2
ventas_tienda2 = tienda2.groupby('Producto').size().sort_values(ascending=False)
most_sold.append(ventas_tienda2.iloc[0])
least_sold.append(ventas_tienda2.iloc[-1])
most_sold_products.append(ventas_tienda2.index[0])
least_sold_products.append(ventas_tienda2.index[-1])

# Tienda 3
ventas_tienda3 = tienda3.groupby('Producto').size().sort_values(ascending=False)
most_sold.append(ventas_tienda3.iloc[0])
least_sold.append(ventas_tienda3.iloc[-1])
most_sold_products.append(ventas_tienda3.index[0])
least_sold_products.append(ventas_tienda3.index[-1])

# Tienda 4
ventas_tienda4 = tienda4.groupby('Producto').size().sort_values(ascending=False)
most_sold.append(ventas_tienda4.iloc[0])
least_sold.append(ventas_tienda4.iloc[-1])
most_sold_products.append(ventas_tienda4.index[0])
least_sold_products.append(ventas_tienda4.index[-1])

# Crear diagrama de dispersión
plt.figure(figsize=(10, 6))
plt.scatter(stores, most_sold, color='green', marker='^', s=100, label='Más vendidos')
plt.scatter(stores, least_sold, color='red', marker='v', s=100, label='Menos vendidos')

# Agregar nombres de productos como anotaciones
for i, (most_prod, least_prod, most_count, least_count) in enumerate(zip(most_sold_products, least_sold_products, most_sold, least_sold)):
    plt.text(i, most_count + 1, most_prod, ha='center', fontsize=8, color='green')
    plt.text(i, least_count - 2, least_prod, ha='center', fontsize=8, color='red')

plt.title('Productos más y menos vendidos por tienda', fontsize=14, pad=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Cantidad de ventas', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.ylim(25, 70)  # Establecer los límites del eje y para centrarse en el rango de ventas
plt.tight_layout()
plt.savefig('product_sales_by_store.png')
```

- Gráficos costo promedio de envío por tienda

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs para las bases de datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar bases de datos
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular el coste medio de envio de cada tienda
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costos_envio = [
    tienda['Costo de envío'].mean().round(2),
    tienda2['Costo de envío'].mean().round(2),
    tienda3['Costo de envío'].mean().round(2),
    tienda4['Costo de envío'].mean().round(2)
]

# Crear gráfico de barras verticales
plt.figure(figsize=(10, 6))
plt.bar(tiendas, costos_envio, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Costo promedio de envío por tienda', fontsize=14, pad=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Costo promedio de envío ($)', fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of bars
for i, v in enumerate(costos_envio):
    plt.text(i, v + 200, f'${v:,.2f}', ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('shipping_cost_by_store.png')
```
